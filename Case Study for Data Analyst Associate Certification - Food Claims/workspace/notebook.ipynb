The original data has **2000** rows and **8** columns. 
There are 36 missing values in `amount_paid` column; 26 missing values in `linked_cases` column. 
I replaced 36 missing values in `amount_paid` with overall median amount paid and 26 missing values in `linked_cases` with FALSE, so there are no missing value.
Also, I replaced 'VEGETABLES' with 'vegetable' and ' Meat' with 'meat' in `cause` column.

I have validated all the columns against the criteria in the dataset table. Let's look at the columns:
- `claim_id`: 2000 unique claim ids, as expected. No cleaning is needed.
- `time_to_close`: positive numeric values without missing values, as expected. No cleaning is needed.
- `claim_amount`: numeric values without missing values, as expected. No cleaning is needed.
- `amount_paid`: numeric values without missing values after cleaning. 36 missing values and replacement has completed.
- `location`: 4 categories without missing values, as expected. No cleaning is needed.
- `individuals_on_claim`: numeric values without missing values, as expected. No cleaning is needed.
- `linked_cases`: boolean values without missing values after cleaning. 26 missing values and replacement has completed.
- `cause`: 3 categories without missing values. Replacement (replace 'VEGETABLES' with 'vegetable', replace ' Meat' with 'meat') is completed.

After the data validation, the dataset contains 2000 rows and 8 columns without missing values.

```python
# import pakcages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
food = pd.read_csv('./food_claims_2212.csv')
```

```python
# Data Validation
# Check if any missing values 
food.isna().sum()
```

```python
food['amount_paid'] = food['amount_paid'].fillna(food['amount_paid'].median())
```

```python
food['linked_cases'] = food['linked_cases'].fillna(False)
```

```python
# Replace `VEGETABLES` with 'vegetable' in 'cause' column
food['cause'] = food['cause'].replace('VEGETABLES', 'vegetable')
```

```python
# Replace ` Meat` with 'meat' in 'cause' column
food['cause'] = food['cause'].replace(' Meat', 'meat')
```
